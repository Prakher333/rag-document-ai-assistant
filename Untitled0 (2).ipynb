```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Intro to Mine legs. and Safety-I_VI Sem_NITRR_Dr. Ashish - Unit 1 (1).pdf")
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vector_db = FAISS.from_documents(
    chunks,
    embeddings
)
```

```python
query = "What is longwall mining?"

docs = vector_db.similarity_search(query)

# Print all retrieved documents to inspect their content
for i, doc in enumerate(docs):
    print(f"--- Document {i+1} ---")
    print(doc.page_content)
    print("\n")
```

```python
print("📄 Mining Document AI Assistant Ready!")
print("Type 'exit' to stop.\n")

while True:
    query = input("Ask your question: ")

    if query.lower() == "exit":
        print("Session ended.")
        break

    # retrieve documents
    docs = vector_db.similarity_search_with_score(query, k=5)

    context = "\n".join([doc.page_content[:500] for doc, score in docs])

    # generate answer
    response = llm(
        question=query,
        context=context
    )

    print("\nAnswer:", response["answer"])
    print("-" * 50)
```
